```python
# CREATING A CUSTOM FUNCTION TO CONCATENATE ALL LOG FILES
# Same function/workflow as the concatenate_logs function above, but now include the ex_rates, sp_rates, and per_species_rates log files in addition to the mcmc log file
# these files should be named with the same prefix as the mcmc log file, but with the suffixes "_sp_rates", "_ex_rates", and "_per_species_rates"
# Save the resulting dataframes as log files to the same directory as the mcmc log file, but with the same prefix as the mcmc log file, and the suffixes "_sp_rates", "_ex_rates", and "_per_species_rates"

import pandas as pd
import numpy as np
import os

def concatenate_logs(original_log_prefix, restored_log_prefix, output_log_prefix):
    suffixes = ['_mcmc.log', '_ex_rates.log', '_sp_rates.log', '_per_species_rates.log']
    
    def process_log_files(original_log_prefix, restored_log_prefix, suffix):
        original_log = original_log_prefix + suffix
        restored_log = restored_log_prefix + suffix
        output_log = output_log_prefix + suffix.replace('.log', '_CONCAT.log')

        # Need to add a "does this file exist" check, b/c  only BDNN runs have the "per_species_rates.log", and I want this function to work on
        # all runs, not just BDNN runs
        if not os.path.exists(original_log) or not os.path.exists(restored_log):
            print(f"Skipping {suffix} because the file does not exist.")
            return None
        
        # Need to add these try and except blocks because RJMCMC sp_rates and ex_rates.logs will have a different number of columns in each row!
        # Which would lead to a parse error when reading in the log file
        try:
            original = pd.read_csv(original_log, sep="\t", header=0)
        except pd.errors.ParserError as e:
            print(f"Skipping {suffix} because of a parsing error in the original log file: {e}")
            return None
        
        try:
            restored = pd.read_csv(restored_log, sep="\t", header=0)
        except pd.errors.ParserError as e:
            print(f"Skipping {suffix} because of a parsing error in the restored log file: {e}")
            return None
      
        # Drop the very first row in the restored log b/c PyRate had an old issue where the first row was restored improperly
        # This issue has been fixed in the current version of PyRate, but the restored logs from the old version still have this issue
        restored = restored.iloc[1:, :]
        concatenated = pd.concat([original, restored], axis=0)
        
        if concatenated.shape[0] == original.shape[0] + restored.shape[0]:
            concatenated_10k = concatenated.iloc[0:10000, :]
            
            if suffix == '_mcmc.log':
                concatenated_10k['it'] = np.arange(0, concatenated_10k.shape[0] * 10000, 10000)
                if concatenated_10k['it'].iloc[-1] != 99990000:
                    raise ValueError(f"The last value in the 'it' column is not 99990000 for {suffix} logs.")
            elif suffix == '_per_species_rates.log':
                concatenated_10k['iterations'] = np.arange(0, concatenated_10k.shape[0] * 10000, 10000)
                if concatenated_10k['iterations'].iloc[-1] != 99990000:
                    raise ValueError(f"The last value in the 'iterations' column is not 99990000 for {suffix} logs.")
            
            concatenated_10k.to_csv(output_log, sep="\t", index=False)
            print(f"Original ({suffix}): {original.shape[0]} rows", 
                  f"Restored ({suffix}): {restored.shape[0]} rows", 
                  f"Concatenated ({suffix}): {concatenated.shape[0]} rows", 
                  f"Concatenated_10k ({suffix}): {concatenated_10k.shape[0]} rows")
            return concatenated_10k
        else:
            raise ValueError(f"The concatenated log file is the wrong length for {suffix} logs.")
    
    results = {}
    for suffix in suffixes:
        result = process_log_files(original_log_prefix, restored_log_prefix, suffix)
        if result is not None: # Results would be none if the file didn't exist (if non-BDNN, with no _per_species_rates.log)
            results[suffix] = result
    return results


# Example Usage:
# Suppose you have the following log files:

# Original log files:
# ./reptilia/mcmc_no_predictors/A_bdnn/reptilia_pyrate_5_G_BDS_BDNN_4_2Tc_mcmc.log
# ./reptilia/mcmc_no_predictors/A_bdnn/reptilia_pyrate_5_G_BDS_BDNN_4_2Tc_ex_rates.log
# ./reptilia/mcmc_no_predictors/A_bdnn/reptilia_pyrate_5_G_BDS_BDNN_4_2Tc_sp_rates.log
# ./reptilia/mcmc_no_predictors/A_bdnn/reptilia_pyrate_5_G_BDS_BDNN_4_2Tc_per_species_rates.log
# Restored log files with the same names but in a different directory.
# You would call the function as follows:

# original_log_prefix = './reptilia/mcmc_no_predictors/A_bdnn/reptilia_pyrate_5_G_BDS_BDNN_4_2Tc'
# restored_log_prefix = './restored_logs/reptilia_pyrate_5_G_BDS_BDNN_4_2Tc'
# restored_log_prefix = './restored_logs/reptilia_pyrate_5_G_BDS_BDNN_4_2Tc'

# results = concatenate_logs(original_log_prefix, restored_log_prefix, output_log_prefix)
# This function will concatenate and process all the specified log files, save the resulting dataframes with the suffix _FINAL.log, and print out the relevant information for each log file. If any check fails, it will raise an error specifying which check failed.
```
