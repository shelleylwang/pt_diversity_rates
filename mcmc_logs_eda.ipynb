## Making a Function

```python
# Read in an incomplete log file
# Read in a restored log file
# Concatenate the two log files
# Check that the concatenated log file is the right length 
# by comparing its shape[0] with the sum of the shape[0]'s of the original and restored log files to the concatenated log file
# Cutting the concatenated log file to 10000 rows
# Renumbering the 'it' column from 0 with step sizes of 10000
# Check that the last value in the 'it' column is 99990000
# Return the resulting dataframe only if all the checks are correct
# return an error message if any of the checks are incorrect, specifying which check was failed
# Save the resulting dataframe as a log file

def concatenate_logs(original_log, restored_log, output_log):
    original = pd.read_csv(original_log, sep="\t", header=0)
    restored = pd.read_csv(restored_log, sep="\t", header=0)
    concatenated = pd.concat([original, restored], axis=0)
    if concatenated.shape[0] == original.shape[0] + restored.shape[0]:
        concatenated_10k = concatenated.iloc[0:10000, :]
        concatenated_10k['it'] = np.arange(0, concatenated_10k.shape[0] * 10000, 10000)
        concatenated_10k.to_csv(output_log, sep="\t", index=False)
        # print out the number of rows in the original, the restored, the concatenated, and the concatenated_10k
        print(f"Original: {original.shape[0]} rows", f"Restored: {restored.shape[0]} rows", 
              f"Concatenated: {concatenated.shape[0]} rows", f"Concatenated_10k: {concatenated_10k.shape[0]} rows")
        return concatenated_10k
    else:
        return "The concatenated log file is the wrong length"
    
```
